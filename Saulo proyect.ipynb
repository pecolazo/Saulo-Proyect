Finally, let's be strict with the only SNs ones that have a single Host and I can't compare them with anything. But first we need to define some functions



```python
def horaGrad(H):
  '''
  returns degrees
  '''
  return(H*15)
def gradHora(G):
  '''
  returns hours
  '''
  return(G/15)

def arcSegGrad(As):  
  '''returns degrees'''
  return(As/3600)

def circulo(ra,ra1,dec,dec1,z,kpc=60):
#'''This function converts two galactic coordinates to Cartesian coordinates on a sphere of radius 1 (for both objects). Then evaluate if the separation (depending on the redshift) is enough to consider the galaxy with a host galaxy'''
  cota=(cosmo.arcsec_per_kpc_proper(z).value)*kpc
  
  if (ra<0):
    phi=(ra+360)*np.pi/180 #ra is in degrate
  else:
    phi=ra*np.pi/180 #ra is in degrate
  
  if (ra1<0):
    phi1=(ra1+360)*np.pi/180
  else:
    phi1=ra1*np.pi/180 #ra is in degrate
  
  theta=(90-gradHora(dec))*np.pi/180 #dec is in Hours
  theta1=(90-gradHora(dec1))*np.pi/180

  x=1*np.sin(theta)*np.cos(phi)
  x1=1*np.sin(theta1)*np.cos(phi1)
  
  y=1*np.sin(theta)*np.sin(phi)
  y1=1*np.sin(theta1)*np.sin(phi1)
  
  z=1*np.cos(theta)
  z1=1*np.cos(theta1)
  
  sep=np.sqrt((x-x1)**2+(y-y1)**2+(z-z1)**2)
  sepcota=arcSegGrad(cota)*np.pi/180 #is in arcsec (3600 arcsec is a 1 degrate)

 # print(sep,sepcota)
  return(sep<sepcota)

```
